```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("stock.csv")
df
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(2, 4, 1)
sns.histplot(df, x="Purchase Price", bins=10, edgecolor='w', color="g")
plt.xlabel('Purchase Price', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.ylabel('Frequency', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.title('Histogram of Purchase Price', fontstyle='italic', fontsize=15, color='g', weight='semibold', family='fantasy')

plt.subplot(2, 4, 2)
sns.histplot(df, x="Current Price", bins=10, edgecolor='w', color="g")
plt.xlabel('Current Price', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.ylabel('Frequency', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.title('Histogram of Current Price', fontstyle='italic', fontsize=15, color='g', weight='semibold', family='fantasy')

plt.subplot(2, 4, 3)
sns.histplot(df, x="Gain/Loss", bins=10, edgecolor='w', color="g")
plt.xlabel('Gain/Loss', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.ylabel('Frequency', fontsize=10, color='g', weight='semibold', family='fantasy')
plt.title('Histogram of Gain/Loss', fontstyle='italic', fontsize=15, color='g', weight='semibold', family='fantasy')

plt.tight_layout()
plt.show()
```

```python
top_dividend = df.nlargest(5, 'Dividend/share')[['Stock Name', 'Symbol', 'Dividend/share']]
print('Top 5 Stocks with highest Dividend/share:')
print(top_dividend)
```

```python
top_market_value = df.nlargest(5, 'Market Value')[['Stock Name', 'Symbol', 'Market Value']]
print('\nTop 5 Stocks with highest Market Value:')
print(top_market_value)
```

```python
top_annual_yield = df.nlargest(7, 'Annual Yield')[['Stock Name', 'Symbol', 'Annual Yield']]
print('\nTop 7 Stocks with highest Annual Yield:')
print(top_annual_yield)
```
